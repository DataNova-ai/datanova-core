```python
# Detectar duplicados
duplicados = df.duplicated()
print(f"Duplicados encontrados: {duplicados.sum()}")

# Eliminar duplicados
df = df.drop_duplicates()
print("Duplicados eliminados.")

# Verificar valores nulos por columna
print("\nValores nulos por columna:")
print(df.isnull().sum())

# Eliminar columnas completamente vacías
df = df.dropna(axis=1, how='all')

# Opcional: Eliminar filas con muchos valores nulos
# df = df.dropna(thresh=int(df.shape[1]*0.7))  # Mantiene filas con al menos 70% de datos válidos

# Rellenar valores nulos (dependiendo del tipo de dato)
# Para columnas numéricas, rellenar con la media
df.fillna(df.select_dtypes(include='number').mean(), inplace=True)

# Para columnas categóricas, rellenar con el valor más frecuente
for col in df.select_dtypes(include='object'):
    df[col].fillna(df[col].mode()[0], inplace=True)

print("\nLimpieza de datos completada ✅")
```

```python
import pandas as pd

# Cargar archivo Excel
df = pd.read_excel("/content/Data Set top peliculas 2019 - Reto Corregido.xlsx")

# Detectar duplicados
duplicados = df.duplicated()
print(f"Duplicados encontrados: {duplicados.sum()}")

# Eliminar duplicados
df = df.drop_duplicates()
print("Duplicados eliminados.")

# Verificar valores nulos por columna
print("\nValores nulos por columna:")
print(df.isnull().sum())

# Eliminar columnas completamente vacías
df = df.dropna(axis=1, how='all')

# Rellenar valores nulos en columnas numéricas con la media
df.fillna(df.select_dtypes(include='number').mean(), inplace=True)

# Rellenar valores nulos en columnas categóricas con la moda
for col in df.select_dtypes(include='object'):
    df[col].fillna(df[col].mode()[0], inplace=True)

print("\nLimpieza completada ✅")
```
