```python
import pandas as pd

# Leer el archivo Excel
df = pd.read_excel("Data Set top peliculas 2019 - Reto Corregido.xlsx")

# Mostrar las primeras filas
print("🔍 Vista previa del dataset:")
display(df.head())

# ----------------------------
# LIMPIEZA DE DATOS
# ----------------------------

# 1. Detección de duplicados
duplicados = df.duplicated()
print(f"🧩 Duplicados encontrados: {duplicados.sum()}")

# 2. Eliminación de duplicados
df = df.drop_duplicates()
print("✅ Duplicados eliminados.")

# 3. Revisión de valores nulos por columna
print("\n🔍 Valores nulos por columna:")
print(df.isnull().sum())

# 4. Eliminar filas con valores nulos
df = df.dropna()
print("\n✅ Filas con valores nulos eliminadas.")

# 5. Confirmar forma final del dataset
print(f"\n✅ Dataset limpio. Dimensiones finales: {df.shape}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Si ya tienes el DataFrame `df` cargado y limpio, no necesitas volver a cargarlo.

# Separar múltiples géneros en columnas individuales (si están separados por comas)
df_exploded = df.copy()
df_exploded['Genres'] = df_exploded['Genres'].str.split(', ')
df_exploded = df_exploded.explode('Genres')

# Contar frecuencia por género
genre_counts = df_exploded['Genres'].value_counts()

# Visualización
plt.figure(figsize=(12, 6))
genre_counts.plot(kind='bar')
plt.title('Distribución de Géneros de Películas')
plt.xlabel('Género')
plt.ylabel('Cantidad de Películas')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
